#5. Derivados

```python
#Función para d1
def d1(S, K, r, desv, T):
  return (np.log(S/K) + (r + desv**2 / 2) * T)/(desv * np.sqrt(T))
```

```python
#Función para d2
def d2(S, K, r, desv, T):
  return (np.log(S/K) + (r - desv**2 / 2) * T)/(desv * np.sqrt(T))
```

```python
#Fórmula de Black - Scholes
def BSF ( S, K, r, desv, T):
  return (S * norm.cdf(d1(S, K, r, desv, T))) - (K * np.exp(-r * T) * norm.cdf(d2(S, K, r, desv, T)))
```
